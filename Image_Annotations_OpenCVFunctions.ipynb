# Image Annotations using OpenCV

In this Jupyter notebook let's try to draw different shapes using **OpenCV**

````
#let's start with the Imports 
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
````

## Read the Image 
* Read the image using **cv2.imread()** function 
* Display the image using **plt.imshow()** function 

* When **cv2.imread()** function reads am image the color scale is in **BGR** format so everytime we are displaying the image we will be converting it back to **RGB** format

````
import requests
url = "https://images.unsplash.com/photo-1534361960057-19889db9621e?ixid=MXwxMjA3fDB8MHxwaG90by1wYWdlfHx8fGVufDB8fHw%3D&ixlib=rb-1.2.1&auto=format&fit=crop&w=640&q=80"
r = requests.get(url)
with open('sample.jpg', 'wb') as f:
    f.write(r.content) 
````

````
img= cv2.imread('sample.jpg')
plt.figure(figsize = [10,10])
plt.imshow(img[:,:,::-1])
````

# Draw a Line

* Let's draw a line using **line()** function from **OpenCV**
* we will drawing a line from one point to another point

````
#To draw the line on the image we are making the copy of the origninal image using copy() function

imageLine = img.copy()
#draw the image from point  A to B
pointA = (250,80)
pointB = (530,80)
cv2.line(imageLine, pointA, pointB, (255, 255, 0), thickness=3, lineType=cv2.LINE_AA)
plt.figure(figsize = [10,10])
plt.imshow(imageLine[:,:,::-1])
````

# Draw a Circle

* we will drawing the circle from one center point with a radius r on the image.
* we will be using **circle()** function from OpenCV
* We will be drawing from a center point which you can define and with an radius **r** 

````
#to draw the circle on the image we are making the copy of the orginal image similar to above

imageCircle = img.copy()

# draw a circle from a point and with a radius 
center = (415,240)
radius =170
cv2.circle(imageCircle, center, radius, (0, 0, 255), thickness=3, lineType=cv2.LINE_AA)

plt.figure(figsize = [10,10])
plt.imshow(imageCircle[:,:,::-1])
````

## Draw a filled circle 

* Now if you want to fill the circle what would you do?
* Actually it's pretty simple you just have to change the thickness to -1


````
imageFilledCircle = img.copy()

# draw a circle from a point and with a radius 
cv2.circle(imageFilledCircle, (415, 240), 170, (255, 0, 0), thickness=-1, lineType=cv2.LINE_AA)

plt.figure(figsize = [10,10])
plt.imshow(imageFilledCircle[:,:,::-1])
````

# Draw a Rectangle

* We will use **rectangle()** function to draw a rectangle on an image.

```
rectangle(img, pt1, pt2, color[, thickness[, lineType[, shift]]]) -> img
```
The mandatory arguments are as follows.
1. `img`: Image on which the rectangle is to be drawn.
2. `pt1`: Vertex of the rectangle. Usually we use the **top-left vertex** here.
3. `pt2`: Vertex of the rectangle opposite to `pt1`. Usually we use the **bottom-right vertex** here.
4. `color`: Rectangle color

````
imageRectangle = img.copy()
start_point = (290,80)
end_point = (520,390)

cv2.rectangle(imageRectangle, start_point, end_point, (0, 0, 255), thickness= 3, lineType=cv2.LINE_8)


plt.figure(figsize = [10,10])
plt.imshow(imageRectangle[:,:,::-1])
````

# Draw an Ellipse 

* You can draw an ellipse by using **ellipse()** function in **OpenCV**
```
ellipse(img, center, axes, angle, startAngle, endAngle, color[, thickness[, lineType[, shift]]]) -> img
```

* You can define the **center** point for the ellipse and **axes** is used for defining the radius of major and minor axes of ellipse

* You can also define the **angle , startAngle and endAngle** for the ellipse so you can plot it in different ways in the image


````
imageEllipse = img.copy()
center = (415,240)
axes1 = (225,165)

axes2=(250,135)
cv2.ellipse(imageEllipse, center, axes1, 0, 0, 360, (255, 0, 0), thickness=3, lineType=cv2.LINE_AA)
cv2.ellipse(imageEllipse, center, axes2, 90, 0, 360, (0, 0, 255), thickness=3, lineType=cv2.LINE_AA)

plt.figure(figsize = [10,10])
plt.imshow(imageEllipse[:,:,::-1])
````

## Draw an Half Ellipse 

* You could draw an Half Ellispe by changing the angles 

````
halfEllipse = img.copy()
# Incomplete/Open ellipse
cv2.ellipse(halfEllipse, (415, 240), (225, 165), 0, 180, 360, (255, 0, 0), thickness=3, lineType=cv2.LINE_AA)
# Filled ellipse
cv2.ellipse(halfEllipse, (415, 240), (225, 165), 0, 0, 180, (0, 0, 255), thickness=-2, lineType=cv2.LINE_AA)

# Incomplete/Open ellipse
#cv2.ellipse(halfEllipse, (415, 190), (100, 50), 0, 180, 360, (255, 0, 0), thickness=3, lineType=cv2.LINE_AA)
# Filled ellipse
#cv2.ellipse(halfEllipse, (415, 190), (100, 50), 0, 0, 180, (0, 0, 255), thickness=-2, lineType=cv2.LINE_AA)

plt.figure(figsize = [10,10])
plt.imshow(halfEllipse[:,:,::-1])
````

# Draw Text

* Let's see how can we write text to the images using **putText()** function in **OpenCV**


```
putText(img, text, org, fontFace, fontScale, color[, thickness[, lineType[, bottomLeftOrigin]]]) -> img
```

* You can define the string in the **text** variable
* **org** is used for placing the text some where over the image 
* You can change the font using **fontFace** and its size using **fontScale**


````
imageText = img.copy()

#let's write the text you want to put on the image
text = 'I am an annotated  dog!'

#org: Where you want to put the text
org = (50,350)

cv2.putText(imageText, text, org, fontFace = cv2.FONT_HERSHEY_COMPLEX, fontScale = 1,
            color = (250,225,100),thickness =  3, lineType=cv2.LINE_AA)

plt.figure(figsize = [10,10])
plt.imshow(imageText[:,:,::-1])
````
