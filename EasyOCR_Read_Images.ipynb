```python
#2
import easyocr
import cv2
from google.colab.patches import cv2_imshow
```

```python
#3
#Instantiate EasyOCR Reader Object
reader = easyocr.Reader(['en'])
```

```python
import cv2
TEST_IMAGE_PATHS = '/content/drive/MyDrive/AI Engineer - ProductizeTech Assignment /TEST IMAGES'
```

```python
result = reader.readtext(TEST_IMAGE_PATHS[3], detail = 1, paragraph = False)

result
```

```python
import numpy as np
for (coord, text, prob) in result:
    (topleft, topright, bottomright, bottomleft) = coord
    tx,ty = (topleft[0], topleft[1])
    bx,by = (bottomright[0], bottomright[1])
    #img = np.ones((100, 100, 3), dtype=np.uint8)
    #img = cv2.cvtColor(TEST_IMAGE_PATHS[3], cv2.COLOR_RGB2BGR)
    ocr_img=cv2.rectangle(cv2.imread(TEST_IMAGE_PATHS[3]), (tx,ty), (bx,by), (0, 255, 0), 2) 
```

```python
Outputmg=cv2.putText(ocr_img,text,(tx, ty), cv2.FONT_HERSHEY_COMPLEX_SMALL, 0, (0, 255, 0), 1)
cv2_imshow(ocr_img)
```
